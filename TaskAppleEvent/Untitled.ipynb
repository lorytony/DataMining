```python
import pandas as pd
import itertools
import collections
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords


stop_words = list(set(stopwords.words('english')))
no_words = ["iphone","apple","youtube","amp","note","tv","even","device","cannot","phone","like","ipad","u","send","one","screen","watch"]
for word in no_words:
    stop_words.append(word)
#print(stop_words)
#df = pd.read_csv("CleanedTweets8to13Sept.csv",index_col=False)
df = pd.read_csv("apple_after_event.csv",index_col=False)
df = df[df['tag']=='negative']

column_text = df["Tweet Text"].tolist()

#print(column_text)

words_in_tweet = [tweet.lower().split() for tweet in column_text]
#print(words_in_tweet)






all_words = list(itertools.chain(*words_in_tweet))
#print(all_words)

all_words_nsw = [word for word in all_words if not word in stop_words]


#print(all_words_nsw)

counts_frequency = collections.Counter(all_words_nsw)

print(counts_frequency.most_common(15))
```

```python
clean_tweets_nsw = pd.DataFrame(counts_frequency.most_common(15),
                             columns=['words', 'count'])

clean_tweets_nsw.at[0,'words'] = "iphone13"
fig, ax = plt.subplots(figsize=(8, 8))

# Plot horizontal bar graph
clean_tweets_nsw.sort_values(by='count').plot.barh(x='words',
                      y='count',
                      ax=ax,
                      color="black")

ax.set_title("Common Words Found in Negative Tweets")

plt.show()
```
